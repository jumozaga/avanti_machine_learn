## 1. Criar um gráfico de linha com títulos nos eixos X e Y:

```python
import matplotlib.pyplot as plt

#Média Geral dos Alunos por Disciplina
disciplinas = ['Matematica', 'Portugues', 'Inglês']
media = [5, 7, 10]

plt.figure(figsize=(9, 3))

plt.subplot(131)
plt.bar(disciplinas, media)
plt.xlabel('Disciplinas')
plt.ylabel('Média')
plt.title(' Barras')


plt.subplot(132)
plt.scatter(disciplinas, media)
plt.xlabel('Disciplinas')
plt.ylabel('Média')
plt.title('Scatter Plot')


plt.subplot(133)
plt.plot(disciplinas, media)
plt.xlabel('Disciplinas')
plt.ylabel('Média')
plt.title('Line Plot')


plt.show()
```

## 2. Usar a função subplot() para mostrar 6 gráficos:

```python
fig, df = plt.subplots(ncols=2, nrows=2, figsize=(9.5, 7.5))

for row in range(2):
    for col in range(2):
        df[row, col].annotate(f'df[{row}, {col}]', (0.5, 0.5),
                               transform = df[row, col].transAxes,
                               ha='center', va='center',
                               fontsize=18, color='darkgrey')
fig.suptitle('Tudo em subplots')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(ncols=2, nrows=2, figsize=(5.5, 3.5),
                        constrained_layout=True)

for row in range(2):
    for col in range(2):
        axs[row, col].annotate(f'axs[{row}, {col}]', (0.5, 0.5),
                               transform=axs[row, col].transAxes,
                               ha='center', va='center',
                               fontsize=18, color='darkgrey')
fig.suptitle('Subplots')
plt.show()
```

## 8. Utilizando pandas, como lidar com valores ausentes (NaN) em um DataFrame?

## 10. Utilizando pandas, mostre uma forma de detectar e tratar valores outliers em um DataFrame.

```python
from scipy import stats

# Detect outliers
z_scores = np.abs(stats.zscore(df['idade']))
outliers = df.index[z_scores > 3]

# Replace outliers with median age
median_age = df['idade'].median()
df.loc[outliers, 'idade'] = median_age

print(df)
```
